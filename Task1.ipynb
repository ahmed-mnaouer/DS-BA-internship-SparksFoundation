```python
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
prediction.describe()
```

```python
plt.title('Scores Distribution Plot')
sns.distplot(prediction['Scores'])
plt.show()
```

```python
plt.scatter(prediction['Hours'], prediction['Scores'], color = 'royalblue')
plt.title('Does Studying Pay Off ?')
plt.xlabel('Hours of Study')
plt.ylabel('Score')
plt.box(False)
plt.show()
```

```python
X = prediction.iloc[:, :1]
Y = prediction.iloc[:, 1:]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
Y_pred_test = regressor.predict(X_test)
Y_pred_train = regressor.predict(X_train)
```

```python
plt.scatter(X_train, Y_train, color = 'royalblue')
plt.plot(X_train, Y_pred_train, color = 'darkblue')
plt.title('Does Studying Pay Off ? (Training Set)')
plt.xlabel('Hours of Study')
plt.ylabel('Score')
plt.legend(['X_train/Pred(Y_test)', 'X_train/Y_train'], title = 'Score/Hours', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, Y_test, color = 'royalblue')
plt.plot(X_train, Y_pred_train, color = 'darkblue')
plt.title('Does Studying Pay Off ? (Test Set)')
plt.xlabel('Hours of Study')
plt.ylabel('Score')
plt.legend(['X_train/Pred(Y_test)', 'X_train/Y_train'], title = 'Score/Hours', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
regressor.fit(X_train, Y_train)
input_data = np.array([[9.25]])
predicted_score = regressor.predict(input_data)
print("Predicted score for 9.25 hours of study:", predicted_score)
warnings.filterwarnings("ignore")
```
